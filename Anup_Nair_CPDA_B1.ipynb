## Q1: Hypothesis Test – Altona

Null: Mean price in Altona = 800000  Alternate1: Mean price in Altona > 80000


```python
altona_prices = df[df['Suburb'] == 'Altona']['Price'].dropna()
```

```python
t_stat = (altona_prices.mean() - 800000) / (
    altona_prices.std(ddof=1) / math.sqrt(len(altona_prices))
)
t_stat
```

AN: The calculated t value is 1.03, which is less than the critical value of 1.645.
Therefore, the null hypothesis is not rejected, and there is insufficient statistical evidence to conclude that the average property price in Altona is greater than $800,000.

## Q2. Summer vs Winter Property Prices (2016)

Null: summer = winter

alternate: summer not equal to winter

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)
df_2016 = df.loc[df['Date'].dt.year == 2016].copy()
df_2016['Month'] = df_2016['Date'].dt.month
```

```python
winter = df_2016[df_2016['Month'].isin([10, 11, 12, 1, 2, 3])]['Price'].dropna()
summer = df_2016[df_2016['Month'].isin([4, 5, 6, 7, 8, 9])]['Price'].dropna()
```

```python
t_stat = (winter.mean() - summer.mean()) / np.sqrt(
    winter.var(ddof=1)/len(winter) +
    summer.var(ddof=1)/len(summer)
)
t_stat
```

AN: An independent two-sample t-test was conducted at the 5% significance level to compare property prices sold during summer and winter months in 2016.
The calculated t-statistic is 3.92, which exceeds the critical value of > 1.96.
Therefore, the null hypothesis is rejected, and there is a statistically significant difference between summer and winter property prices in 2016.
